#  Environment and Project Setup

This notebook checks the runtime environment, mounts Google Drive, and verifies the main project root and raw data folder.

## Check Python environment

```python
import sys
import os

print("Python environment")
print(f"Python version: {sys.version}")
print(f"Python executable: {sys.executable}")
print(f"Current working directory: {os.getcwd()}")
```

## Detect runtime environment

```python
# Detect whether running in Google Colab
try:
    import google.colab
    IN_COLAB = True
    print("Runtime: Google Colab")
except:
    IN_COLAB = False
    print("Runtime: local")

print(f"IN_COLAB = {IN_COLAB}")
```

## Mount Google Drive

```python
if IN_COLAB:
    print("Mounting Google Drive...")
    from google.colab import drive
    drive.mount('/content/drive')
    print("Google Drive mounted")
else:
    print("Skipping Google Drive mount (not in Colab)")
```

## Set project root

```python
# Set project root
if IN_COLAB:
    PROJECT_ROOT = "/content/drive/MyDrive/breast_cancer_detection"
else:
    PROJECT_ROOT = "C:/Users/Farid/Documents/MLProjects"

print(f"PROJECT_ROOT: {PROJECT_ROOT}")
```

## Check project root

```python
print("Checking project root...")

if os.path.exists(PROJECT_ROOT):
    print("Project root found")
    print(f"Path: {PROJECT_ROOT}")
else:
    print(f"Project root not found: {PROJECT_ROOT}")
```

## Check raw_data folder

```python
print("Checking raw_data folder...")

if not os.path.exists(PROJECT_ROOT):
    print(f"Project root not found: {PROJECT_ROOT}")
else:
    raw_data_path = os.path.join(PROJECT_ROOT, "raw_data")
    if os.path.exists(raw_data_path):
        print("raw_data folder found")
        print(f"Path: {raw_data_path}")
    else:
        print(f"raw_data folder not found: {raw_data_path}")
```

## Check Kaggle credentials file

```python
print("Checking kaggle.json...")

if not os.path.exists(PROJECT_ROOT):
    print(f"Project root not found: {PROJECT_ROOT}")
else:
    kaggle_json_path = os.path.join(PROJECT_ROOT, "kaggle.json")
    if os.path.exists(kaggle_json_path):
        print("kaggle.json found")
        print(f"Path: {kaggle_json_path}")
    else:
        print(f"kaggle.json not found: {kaggle_json_path}")
```

## Final setup validation

```python
def validate_project_setup(project_root):
    """Validate project root and raw_data folder."""

    print("Final setup validation")

    root_exists = os.path.exists(project_root)
    raw_data_path = os.path.join(project_root, "raw_data")
    raw_data_exists = os.path.exists(raw_data_path)

    if root_exists:
        print(f"PROJECT_ROOT exists: {project_root}")
    else:
        print(f"PROJECT_ROOT not found: {project_root}")

    if raw_data_exists:
        print(f"raw_data exists: {raw_data_path}")
    else:
        print(f"raw_data not found: {raw_data_path}")

    is_ready = root_exists and raw_data_exists
    print(f"Validation result: {'ready' if is_ready else 'not ready'}")

    return is_ready

# Run validation
is_ready = validate_project_setup(PROJECT_ROOT)
```
