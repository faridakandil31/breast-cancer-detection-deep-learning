# Metadata and Label Preparation

This notebook prepares the metadata used to train the model.  
The pathology field is converted into binary labels so the images can be used for classification.

```python
import os
import pandas as pd
import numpy as np

print("Libraries imported")
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
```

## Inspect metadata
Set project paths for metadata files and images.

```python
# Define project paths
PROJECT_ROOT = "/content/drive/MyDrive/breast_cancer_detection"
RAW_DATA_PATH = os.path.join(PROJECT_ROOT, "raw_data")
CSV_PATH = os.path.join(RAW_DATA_PATH, "csv")
JPEG_PATH = os.path.join(RAW_DATA_PATH, "jpeg")


print("PROJECT PATHS")

print(f"Project Root:  {PROJECT_ROOT}")
print(f"Raw Data:      {RAW_DATA_PATH}")
print(f"CSV Metadata:  {CSV_PATH}")
print(f"JPEG Images:   {JPEG_PATH}")
```

## Inspect metadata
Check required folders and metadata CSV files.

```python
print("CHECKING FOLDERS AND FILES")

# Check folders
folders_to_check = [
    ("Raw Data", RAW_DATA_PATH),
    ("CSV Metadata", CSV_PATH),
    ("JPEG Images", JPEG_PATH)
]

print("\nFolders:")
all_folders_exist = True
for name, path in folders_to_check:
    exists = os.path.exists(path)
    status = "OK" if exists else "MISSING"
    print(f"  {status} {name}: {path}")
    if not exists:
        all_folders_exist = False

# Check CSV files
csv_files = [
    "mass_case_description_train_set.csv",
    "mass_case_description_test_set.csv",
    "calc_case_description_train_set.csv",
    "calc_case_description_test_set.csv"
]

print("\nCSV files:")
all_files_exist = True
for csv_file in csv_files:
    file_path = os.path.join(CSV_PATH, csv_file)
    exists = os.path.exists(file_path)
    status = "OK" if exists else "MISSING"
    print(f"  {status} {csv_file}")
    if not exists:
        all_files_exist = False


if all_folders_exist and all_files_exist:
    print("All required folders and files found")
else:
    print("Some folders or files are missing")
    print("Check dataset extraction")
```

## Load metadata
Load metadata CSV files into dataframes.

```python
print("LOADING CSV FILES")

# Load each CSV file
mass_train = pd.read_csv(os.path.join(CSV_PATH, "mass_case_description_train_set.csv"))
print("Loaded: mass_case_description_train_set.csv")

mass_test = pd.read_csv(os.path.join(CSV_PATH, "mass_case_description_test_set.csv"))
print("Loaded: mass_case_description_test_set.csv")

calc_train = pd.read_csv(os.path.join(CSV_PATH, "calc_case_description_train_set.csv"))
print("Loaded: calc_case_description_train_set.csv")

calc_test = pd.read_csv(os.path.join(CSV_PATH, "calc_case_description_test_set.csv"))
print("Loaded: calc_case_description_test_set.csv")

print("\nAll CSV files loaded")
```

## Prepare labels
Combine metadata tables and keep source information.

```python
print("COMBINING ALL CSV FILES")

# Add source_file column to each dataframe
mass_train['source_file'] = 'mass_train'
mass_test['source_file'] = 'mass_test'
calc_train['source_file'] = 'calc_train'
calc_test['source_file'] = 'calc_test'

print("Added 'source_file' column to each dataframe")

# Concatenate all dataframes
combined_df = pd.concat([
    mass_train,
    mass_test,
    calc_train,
    calc_test
], ignore_index=True)

print("Concatenated all dataframes")
print("Combined dataframe created")
```

## Check label distribution
Inspect pathology label counts used for classification.

```python
print("PATHOLOGY DISTRIBUTION")

if 'pathology' in combined_df.columns:
    print("\nDiagnosis label counts:")
    pathology_counts = combined_df['pathology'].value_counts()
    print(pathology_counts)
    
    print("\nPercentage distribution:")
    pathology_pct = combined_df['pathology'].value_counts(normalize=True) * 100
    for label, pct in pathology_pct.items():
        print(f"  {label}: {pct:.2f}%")
else:
    print("\n'pathology' column not found in dataframe")
```

## Inspect metadata
Check BI-RADS assessment value distribution.

```python
print("ASSESSMENT DISTRIBUTION (BI-RADS)")

if 'assessment' in combined_df.columns:
    print("\nAssessment category counts:")
    assessment_counts = combined_df['assessment'].value_counts().sort_index()
    print(assessment_counts)
    
    print("\nBI-RADS categories:")
    print("  0 = Incomplete")
    print("  1 = Negative")
    print("  2 = Benign")
    print("  3 = Probably Benign")
    print("  4 = Suspicious")
    print("  5 = Highly Suggestive of Malignancy")
else:
    print("\n'assessment' column not found in dataframe")
```

## Prepare labels
Convert the pathology values into binary labels so the dataset can be used for a binary classification model (benign vs malignant).

```python
print("Label column identification")

if 'pathology' in combined_df.columns:
    print("Label column found: pathology")

    # Clean pathology text
    pathology_clean = combined_df['pathology'].astype(str).str.strip().str.upper()

    # Keep valid classes
    valid_mask = pathology_clean.str.contains('BENIGN|MALIGNANT', na=False)
    filtered_df = combined_df.loc[valid_mask].copy()
    pathology_clean = pathology_clean.loc[valid_mask]

    # Convert pathology to binary label (0=benign, 1=malignant)
    filtered_df['label'] = pathology_clean.apply(lambda x: 1 if 'MALIGNANT' in x else 0)

    # Save cleaned dataframe
    combined_df = filtered_df

    print(f"Rows after label filtering: {len(combined_df)}")
    print("Binary label mapping: BENIGN -> 0, MALIGNANT -> 1")

    missing_labels = combined_df['label'].isna().sum()
    print(f"Missing values in label: {missing_labels}")
else:
    print("Error: 'pathology' column not found")
    print("Available columns:")
    for col in combined_df.columns:
        print(f"- {col}")
```

## Check label distribution
Summarize metadata size and label counts.

```python
print("Metadata summary")

print(f"Total rows: {combined_df.shape[0]:,}")
print(f"Total columns: {combined_df.shape[1]}")

print("Dataset composition:")
for source in combined_df['source_file'].unique():
    count = (combined_df['source_file'] == source).sum()
    print(f"- {source}: {count:,}")

if 'pathology' in combined_df.columns:
    print("Pathology distribution:")
    for label in combined_df['pathology'].unique():
        count = (combined_df['pathology'] == label).sum()
        pct = (count / len(combined_df)) * 100
        print(f"- {label}: {count:,} ({pct:.1f}%)")

if 'label' in combined_df.columns:
    print("Binary label distribution:")
    print(combined_df['label'].value_counts().sort_index())
```
